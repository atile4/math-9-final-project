```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_raw = pd.read_csv("../student_data_categorized.csv")
```

```python
df_raw.head()
```

```python
df_raw.columns
```

```python
# Features

print("Number of rows:", len(df_raw))
print()
```

```python
# Plotting Grade Distribution
sns.histplot(df_raw['1st_sem_grade'], bins=20)
plt.xlabel("Average Grade")
plt.ylabel("Number of Students")
plt.title("Average Grade Distribution in 1st Semester (Max 20)")
plt.show()
```

```python
# Plotting Grade Distribution
sns.histplot(df_raw['2nd_sem_grade'], bins=20)
plt.xlabel("Average Grade")
plt.ylabel("Number of Students")
plt.title("Average Grade Distribution in 2nd Semester (Max 20)")
plt.show()
```

```python
# Plotting Grade Distribution
sns.histplot(df_raw['Age at enrollment'], bins=10)
plt.xlabel("Age")
plt.ylabel("Number of Students")
plt.title("Age at Enrollment Distribution")
plt.show()
```

```python
ax = sns.countplot(x='Course', data=df_raw)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=8)
plt.tight_layout()
plt.title("Distribution of courses")
plt.show()
```

```python
ax = sns.countplot(x='Target', data=df_raw)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=8)
plt.tight_layout()
plt.title("Distribution of Target")
plt.show()
```
