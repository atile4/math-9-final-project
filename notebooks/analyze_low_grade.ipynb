# Analyzing features of students with a 0 grade

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../student_data_categorized.csv")
```

```python
print("Fraction of students who recieved a 0 grade 1st sem:", len(df[df["1st_sem_grade"] == 0]), "/", len(df))
print("Fraction of students who recieved a 0 grade 2nd sem:", len(df[df["2nd_sem_grade"] == 0]), "/", len(df))
```

```python
df_zero = df[(df["1st_sem_grade"] == 0) | (df["2nd_sem_grade"] == 0)]
df_zero.head()
```

```python
ax = sns.countplot(x='Previous qualification', data=df_zero)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=8)
plt.tight_layout()
plt.title("Distribution of previous qualifications")
plt.show()
```

```python
ax = sns.countplot(x='Target', data=df_zero)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right', fontsize=8)
plt.tight_layout()
plt.title("Distribution of Target (Zero grade)")
plt.show()
```

Looks like proportion of students with a zero grade have a significantly higher dropout, for obvious reasons.

```python
ax = sns.countplot(x='Daytime/evening attendance', data=df_zero)
ax.set_xticklabels(ax.get_xticklabels())
plt.tight_layout()
plt.title("Distribution of daytime/evening attendance")
plt.show()
```

```python
ax = sns.countplot(x='Daytime/evening attendance', data=df)
ax.set_xticklabels(ax.get_xticklabels())
plt.tight_layout()
plt.title("Distribution of daytime/evening attendance")
plt.show()
```

Looks like daytime/nighttine attendance doesn't have much effect
